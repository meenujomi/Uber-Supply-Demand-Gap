## 1.3 Cleaning the dataframe

<font color= green>___From the above .info() it is visible that the dates and time column are not in the right format___</font>

```python
# Converting the last two columns into datetime
uber_df['Request timestamp'] = pd.to_datetime(uber_df['Request timestamp'], dayfirst = True)
uber_df['Drop timestamp'] = pd.to_datetime(uber_df['Drop timestamp'], dayfirst = True)
uber_df.info()
```

```python
# To find the number of null values for each column
uber_df.isnull().sum()
```

```python
# To check the different status availabe
uber_df['Status'].drop_duplicates(keep= 'first')
```

<font color= green>___There are many null values in `drop timestamp`. This could be because the `status` of them are 'Cancelled' or 'No cars available'___</font>

```python
uber_df.loc[uber_df['Status'] == 'Trip Completed'].isnull().sum()
```

```python
uber_df.loc[uber_df['Status'] != 'Trip Completed'].isnull().sum()
```

<font color= green>___There are no drop off time in the the above since those cabs were not taken.___</font>

```python
uber_df.loc[uber_df['Status'] == 'No Cars Available'].isnull().sum()
```

<font color= green>___There are many null values in `Driver id`. This is because there were 'No cars available'___</font>

<font color= green>___We can convert everything into lower for easy accessibility___</font>

```python
uber_df['Pickup point'] = uber_df['Pickup point'].str.lower()
uber_df['Status'] = uber_df['Status'].str.lower()
uber_df
```

<font color= green>___Making different columns that we will require later for the analysis___</font>

```python
# For Request date
uber_df['Request Date'] = uber_df['Request timestamp'].dt.date
# For Request time
uber_df['Request Time'] = uber_df['Request timestamp'].dt.time
# For the Day
uber_df['Day'] = uber_df['Request Date'].apply(lambda x : x.day)
# For Drop date
uber_df['Drop Date'] = uber_df['Drop timestamp'].dt.date
# For Drop time
uber_df['Drop Time'] = uber_df['Drop timestamp'].dt.time
uber_df
```

<font color= green>___Making different time slots___</font>

```python
# For Pick up time slots
uber_df['Request Hour'] = uber_df['Request timestamp'].dt.hour
uber_df['Request Slot'] = pd.cut(uber_df['Request Hour'], bins=[0,1,4,7,10,13,16,18,21,23,24], labels = ['Late Night','Early Morning','Morning','Late Morning','Early Afternoon','Afternoon','Evening','Late Evening','Night','Late Night 1'], right= False)
uber_df['Request Slot'] = uber_df['Request Slot'].replace('Late Night 1', 'Late Night')
uber_df = uber_df.drop(columns = 'Request Hour')
# For Drop time slots
uber_df['Drop Hour'] = uber_df['Drop timestamp'].dt.hour
uber_df['Drop Slot'] = pd.cut(uber_df['Drop Hour'], bins=[0,1,4,7,10,13,16,18,21,23,24], labels = ['Late Night','Early Morning','Morning','Late Morning','Early Afternoon','Afternoon','Evening','Late Evening','Night','Late Night 1'], right= False)
uber_df['Drop Slot'] = uber_df['Drop Slot'].replace('Late Night 1', 'Late Night')
uber_df = uber_df.drop(columns = 'Drop Hour')
uber_df
```

```python
# To check if 'Request id' is not repeated
uber_df['Request id'].drop_duplicates().count()
```

```python
# To check if the Request Date and Drop Date are the same
print(uber_df['Request Date'].isin( uber_df['Drop Date']).sum())
print(uber_df['Request Date'].count())
```

<font color= green>___From above, we can understand that the request and drop date are the same. There will be changes only in time then.___</font>

```python
# To check if the months and year in the data are same
print((uber_df['Request Date'].apply(lambda x : x.month) == 7).count())
print((uber_df['Request Date'].apply(lambda x : x.year) == 2016).count())
print(uber_df['Request Date'].count())
```

<font color= green>___Since all the values come upto 6745 we can confirm that this is the data for the month of July in 2016 and we can drop one of the Date columns___</font>

```python
uber_df = uber_df.drop(columns = 'Drop Date')
uber_df['Date'] = uber_df['Request Date']
uber_df = uber_df.drop(columns = 'Request Date')
uber_df
```

<font color= green>___Now that the data is ready we can move on to Univariate Analysis___</font>

# Step 4 : Bivariate Analysis <br />
 We will take the Variables against each other for this step to understand how much each of them are correlated.<br/>Since all are 'Categorical Variables', we can take the count of each column and correlate them. (Using `Request id` as the column for counting since it is unique throughout) <br/> Heatmap() is used to do the following plots as they will give a good view when we relate two variables with each other. <br/> The colour scheme will help us to know the major changes since they are gradient in these heat maps

 ##  4.1 'Pickup point' & 'Status'

```python
pickup_status = pd.pivot_table(uber_df, values= 'Request id',columns='Pickup point', index = 'Status', aggfunc='count')
pickup_status
```

```python
sns.heatmap(pickup_status, annot = True, fmt='d',cmap= sns.cubehelix_palette(200))
plt.show()
```

# Step 5 : Derived Metrics

 ##  5.1 Gap between 'Supply' and 'Demand' (Business driven Variable)

```python
# Creating a new column named 'Gap' to plot graphs 
uber_df['Gap'] = uber_df['Status'].apply(lambda x: 'Supply' if x == 'trip completed' else 'Demand')
uber_df
```

```python
# Countplot() is used since it is a categorical variable and gradient is given to show which is highest
sns.countplot(uber_df['Gap'],palette =sns.light_palette("green", reverse= True),order = uber_df['Gap'].value_counts().index)
plt.show()
```

<font color= green>___Looks like there is a huge 'Demand' and the 'Supply' is very low___</font>

```python
# Multiple colors are used since we have to show them sepeartly
request = sns.countplot(uber_df['Request Slot'], hue=uber_df['Gap'])
plt.legend(loc='right', bbox_to_anchor=(1.252,.93))
request.tick_params(axis='x', rotation=90)
plt.show()
```

<font color= green>___The highest `Gap` is in the 'Late Evening'___</font>

```python
sns.countplot(uber_df['Pickup point'], hue=uber_df['Gap'])
plt.legend(loc='right', bbox_to_anchor=(1.252,.93))
request.tick_params(axis='x', rotation=90)
plt.show()
```

<font color= green>___It  is clear that the difference is the most in 'Airport' although there is much greater demand in the 'City'___</font>
